```python
class Solution:
    
    def getMaximumGold(self,grid: list[list[int]]) -> int:
            a = self.runner_function(grid)
            
            #test.overall_result = []
            return a
        
        
    def runner_function(self,a):
        obj = test(a,0,0)
        x,y = obj.getdimensions(a)
        for i in range(x):
            for j in range(y):

                if a[i][j] != 0 and (i,j) != (0,0):

                    obj = test(a,i,j)
        return max(obj.overall_result)


class test:
    
    overall_result = []
    
    def __init__(self,original_lst,row,col):
        
        
        self.original_lst = original_lst
        self.stack = []
        #     left = (0,-1)
        #     right = (0,1)
        #     up = (-1,0)
        #     down = (1,0)
        
        self.tempstack = []
        self.results = []
        self.stack.append((row,col))
        self.possible_move = [(0,1),(1,0),(0,-1),(-1,0)]
        self.new_lst =  list(map(list, self.original_lst))
        
        # 4 by 2
        self.m,self.n = self.getdimensions(self.original_lst)
        
        
        self.row = row
        self.col = col
        self.dfs(self.row,self.col,self.new_lst,self.original_lst)
        
        
    def getdimensions(self,lst):
    
        m = len(lst) 
        n = len(lst[0])

        return (m,n)

        
    def dfs(self,r,c,new_lst , original_lst):
        
        
        if len(self.stack)== 0:
            
            test.overall_result.append(max(self.results))
            #self.results = []
            
            
            #print(max(test.overall_result))
            return max(test.overall_result)
        else:
            
            
            new_lst[r][c] = '*'
            new_r,new_c = self.get_coordinates(r,c,new_lst)
            
            
            
            
            
            if (new_r,new_c) == ('r','r'):
                
                
                
                while self.isnext(new_lst,r,c) != True and len(self.stack)!=0:
                    
                    
                    
                    try:
                        x,y,r,c = self.backtracking()
                    except:
                        x,y = self.stack.pop()
                        self.tempstack.append((x,y))
                        
                        break
                    
                    
                    
                    new_r,new_c = self.get_coordinates(r,c,new_lst)
                    
                    
                    self.tempstack.append((x,y))
                    
                    
                    

                    
                
            for i in self.tempstack:
                x,y = i
                new_lst[x][y] = original_lst[x][y]
                
            if (new_r,new_c) != ('r','r'):
                
                self.stack.append((new_r,new_c))
                
            #recursive call
            self.dfs(new_r,new_c,new_lst,original_lst)
    def get_coordinates(self,r,c,new_lst):
        
        cor = self.next_max(new_lst,r,c)
        if cor != (r,c) and cor != 0:
            #print(cor)
            return cor
        
        for i in self.possible_move:
            
            
            
            #self.possible_move = [(1,0),(0,-1),(0,1),(-1,0)]
            x = r+(i[0])
            y = c+(i[1])
            
            if x < 0 or x == self.m or y < 0 or y == self.n or new_lst[x][y] == '*' or new_lst[x][y] == 0 or (x,y) in self.tempstack:
                #yaha back tracking implement hogi ya next coordinate check karna hoga
                x = r
                y = c
            else:
                
                
                return (x,y)
            
        if (x,y) == (r,c):
            
            return ('r','r')

    def next_max(self,new_lst,r,c):
        possible_max = []
        coordinate = 0
        for i in self.possible_move:
            x = r+(i[0])
            y = c+(i[1])
            
            if x < 0 or x == self.m or y < 0 or y == self.n or new_lst[x][y] == '*' or new_lst[x][y] == 0 or (x,y) in self.tempstack:
                
                x = r
                y = c
                
            elif coordinate == 0 and (x,y) != (r,c):
                
                coordinate = (x,y)
                
            elif new_lst[x][y] >= new_lst[coordinate[0]][coordinate[1]] and (x,y) != (r,c):
                possible_max.append((x,y))
                coordinate = (x,y)
       
        return coordinate
            
                
    
    def isnext(self,new_lst,r,c):
        
        cor = self.next_max(new_lst,r,c)
        
        if cor != (r,c) and cor != 0:
            return True
        
        for i in self.possible_move:
            
            
            x = r+(i[0])
            y = c+(i[1])
            
            if x < 0 or x == self.m or y < 0 or y == self.n or new_lst[x][y] == '*' or new_lst[x][y] == 0 or (x,y) in self.tempstack:

                
                
                x = r
                y = c
            else:
                
                
                return True
        
        
        
        return False 
        
#         try:
#             for i in self.possible_move:
            
#                 #self.possible_move = [(1,0),(0,-1),(0,1),(-1,0)]
#                 x = r+(i[0])
#                 y = c+(i[1])
                
#                 new_lst[x][y]
                
#                 if new_lst[x][y] !='*' and new_lst[x][y] != 0:
                    
#                     print('isnext',x,y)
#                     return True
#         except:
#             return False
    def backtracking(self):
        
        
        h = 0
        for i,j in self.stack:
            
            h = h + self.original_lst[i][j]
        
             
            self.results.append(h)
            #print('result ',self.results)
        
        if len(self.stack)> 1:
            
            
            x,y = self.stack.pop()
            prev_r,prev_y = self.stack.pop()
            
            if (prev_r,prev_y) != (self.row,self.col):
                self.stack.append((prev_r,prev_y))
            
            
            return (x,y,prev_r,prev_y)
        else:
            
            
            return False


obj = Solution()

grid =[[34,0,1,0,0,0],[0,0,2,0,1,0],[5,4,3,7,4,2],[0,0,5,0,1,4],[0,0,5,0,2,3]]

(obj.getMaximumGold(grid))

```
